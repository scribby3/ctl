```python
import pandas as pd
```

```python
ppl = pd.read_csv('people.csv') 
```

```python
ppl.head()
```

```python
calls = pd.read_csv('calls.csv') 
calls = calls.rename(columns={'person_ID': 'person_id'})
```

```python
calls.head()
```

```python
single_table = calls.merge(ppl, how="left")
```

```python
single_table = pd.concat([single_table, pd.get_dummies(single_table['Referal Soure'])], axis=1)

single_table
```

```python
single_table["take"].unique()
```

```python
single_table.loc[single_table['take'] == 'yes', 'take'] = 1
single_table.loc[single_table['take'] == 'Yes', 'take'] = 1
single_table.loc[single_table['take'] == 'y', 'take'] = 1
single_table.loc[single_table['take'] != 1, 'take'] = 0
```

```python
final_table = single_table[["take","Facebook","Google","Other","referal","website"]]
```

```python
final_table
```

```python
final_table.to_csv("dewey-cheetham-howe.csv", index_label=False)
```
